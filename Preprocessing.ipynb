# **Preprocessing**

We need to drop the Unnamed columns from dataset, because they are useless.

```python
# Drop Unnamed columns
df.drop(df.columns[df.columns.str.contains('Unnamed',case = False)],axis = 1, inplace = True)
df.head()
```

```python
# Check missing values
df.isna().sum()
```

Based on above exploration, there is no missing values in all the columns.

```python
# Apply lambda function to each row of dataframe to delete ";" character
df = df.apply(lambda row: row.str.strip().str.rstrip(';'))
df.tail()
```

The lambda function uses the str.strip() method to remove any whitespace and other trailing characters from the row, and the str.rstrip() method to remove the semicolon character from the end of the row.

Note: I used the str.rstrip() method instead of the [:-1] string slicing syntax, because the str.rstrip() method removes all trailing characters that match the specified character or characters, which is more robust than simply removing the last character of the string.




```python
# Delete "'" from both sides of each element in the dataframe
df = df.applymap(lambda x: x.strip("'"))
df.head()
```

## **Subtask1**

As is clear, the information in the Date column is not correct. Because based on the project's summary, this dataset is for date between 2022 to July 2033. But, the year for all the Date columns is 2022! So, we need to solve this problem. 

---



**Solution:**


> When I tried to change the column to datatime type, I received this error: ValueError: time data 2022-02-29 doesn't match format specified.
The ValueError I encountered is caused by attempting to convert a date string that is not a valid date, specifically a date that does not exist in the Gregorian calendar, into a datetime object.
In this case, the date string '2022-02-29' is causing the error, because 2022 is not a leap year, and therefore February 29th does not exist in that year.
To handle this error, we can set the errors argument of the pd.to_datetime function to 'coerce', which will convert any invalid date strings to a NaT (Not a Time) value.



> I defined a start date of 2022-01-01 and loops through a range of years from 2022 to 2033. Within the loop, the code checks if the current row of df1 has a Date value equal to the start date. If it does, the year component of the Date value is updated to the current year being looped through.
The code then enters a while loop that continues until it encounters a row where the Date value is equal to the start date. Within this while loop, the year component of each Date value is updated to the current year being looped through. If the end of the DataFrame is reached before encountering a row where the Date value is equal to the start date, the year component of the final Date value is updated to the current year.


```python
# Make a copy of dataframe for next changes
df1 = df

# Convert Date column to datetime format and handling errors
df1['Date'] = pd.to_datetime(df1['Date'], format='%Y-%m-%d', errors='coerce')

i=0
# Define the point which we need to add one to the previous year
start_date = pd.to_datetime('2022-01-01')

for year in range(2022, 2034):
    if (df1.loc[(i,'Date')] == start_date):
      df1.loc[(i, 'Date')] = df1.loc[(i,'Date')].replace(year=df1.loc[(i,'Date')].year + (year-2022))
      i+=1
      while (df1.loc[(i,'Date')] != start_date): # and i<3902
        # if i < 3902:
          df1.loc[(i, 'Date')] = df1.loc[(i,'Date')].replace(year=df1.loc[(i,'Date')].year + (year-2022))  
          i+=1
          if i == len(df)-1:
            df1.loc[(i, 'Date')] = df1.loc[(i,'Date')].replace(year=df1.loc[(i,'Date')].year + (year-2022))
            break  
df1
```
